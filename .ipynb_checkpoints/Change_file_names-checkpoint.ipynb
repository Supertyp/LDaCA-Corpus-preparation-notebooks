> #### <i class="fa fa-gear fa-spin fa-2x" style="color: firebrick"></i> Configuration
> As given, the notebook is configured to make the following changes:<br><br>
> **_anon -> (removed)**<br>
> **AAE_ -> AusESL_**
> <br>

 


```python
import os

# Example path to folder
directory_path = 'C:/Users/barth/Documents/LDACA/AusESL/edited_elan'



def rename_files_in_bulk(directory_path, old_string, new_string):
    try:
        # Iterate through files in the directory
        for filename in os.listdir(directory_path):
            old_file_path = os.path.join(directory_path, filename)

            # Check if the file starts with the specified old prefix
            if old_string in filename:
                new_filename = filename.replace(old_string, new_string)
                new_file_path = os.path.join(directory_path, new_filename)

                # Rename the file
                os.rename(old_file_path, new_file_path)
                print(f"Renamed: {filename} to {new_filename}")

    except Exception as e:
        print(f"Error renaming files: {e}")


# 2 USE CASES:

# remove _anon suffix
old_string = '_anon'
new_string = ''
rename_files_in_bulk(directory_path, old_string, new_string)


# replace AAE_ prefix wit AusESL_ prefix
old_string = 'AAE_'
new_string = 'AusESL_'
rename_files_in_bulk(directory_path, old_string, new_string)


print ("+++ DONE +++")
```
